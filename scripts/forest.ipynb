```python
f_select
```

```python
def standard(arr):
    import numpy as np
    sd = np.std(arr, ddof = 1)
    m = np.mean(arr)
    arr = [(a - m)/sd for a in arr]
    return arr
```

```python
f_std = pd.DataFrame(columns = f_select.columns)
f_std['日期'] = f_select['日期']
```

```python
for name in list(f_select.columns):
    if name != '日期':
        f_std[name] = standard(f_select[name])
```

```python
f_std
```

## 送入模型

```python
import sklearn 
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_predict
from sklearn import metrics

from sklearn import tree
from sklearn import ensemble
```

```python
f_std['y'] = f_std['備轉容量(MW)'].shift(-1)
f_std.fillna(method='pad',inplace=True)
```

```python
f_std
```

## 前兩個是以年為單位做切割，後兩個是以train數量做切割

```python
train = f_std[[str(f_std['日期'][i])[:4] == '2019' for i in range(len(f_std))]]
```

```python
test = f_std[[str(f_std['日期'][i])[:4] == '2020' for i in range(len(f_std))]]
```

```python
#model = linear_model.ElasticNet()
model = ensemble.RandomForestRegressor(n_estimators=100, criterion='mse',max_depth=3)
```

```python
x_list = ['備轉容量(MW)']
```

```python
x = train[x_list].values
y = train['y'].values
```

### 用多項式試試看

```python
model.fit(x,y)
```

```python
xtest = test[x_list].values

# xtest0 = test[x_list].values
# xtest = poly.fit_transform(xtest0)


test['y2'] = model.predict(xtest)
```

```python
test
```

```python
ytest = test['y'].values
y2 = test['y2'].values
```

```python
err = metrics.mean_squared_error(ytest, y2)
err
```
